```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
from datetime import datetime
import matplotlib.dates as mdates
```

```python
df = pd.read_csv("province.csv")
df.head()
```

```python
df = pd.read_csv("daily.csv")
df
```

```python
pd.to_datetime(df["date"].values)
```

```python
date = np.array(pd.to_datetime(df["date"].values))
nowConfirm = df["nowConfirm"].values
importedCase = df["importedCase"].values
dead = df["dead"].values
confirm = df["confirm"].values
heal = df["heal"].values
date
```

```python
date = df["date"].values
date_list = list(date)
for i in range(len(date_list)):
    date_list[i] = datetime.strptime(date_list[i], '%Y-%m-%d').date()
#datetime.strptime(str, '%Y/%m/%d').date_list()
#date_list
```

```python
plt.plot(date_list,nowConfirm, color="g")
plt.xlabel("date")
plt.ylabel("nowConfirm")
plt.gcf().autofmt_xdate()
plt.show()
```

```python
plt.plot(date_list,importedCase)
plt.xlabel("date")
plt.ylabel("importedCase")
plt.gcf().autofmt_xdate()
plt.show()
```

```python
plt.plot(date_list,dead,label="dead")
plt.plot(date_list,confirm,label="confirm")
plt.plot(date_list,heal,label="heal")
plt.legend(loc=0,ncol=1)
plt.xlabel("date")
plt.ylabel("num")
plt.gcf().autofmt_xdate()
plt.show()
```
