```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1, 1)
y = 4 * x**2 + x + 2 + np.random.normal(0, 1, 100)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X, y)
y_pred = lr.predict(X)
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred, color = 'r')
plt.show()
```

```python
X2 = np.hstack([X, X ** 2, X ** 3, X ** 4])
X2[:3]
```

```python
X2.shape
```

```python
lr.fit(X2, y)
y_pred = lr.predict(X2)
```

```python
plt.scatter(x, y)
plt.plot(np.sort(x), y_pred[np.argsort(x)], color = 'r')
plt.show()
```

```python
lr.coef_
```

```python
x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1, 1)
y = 4 * x**2 + x + 2 + np.random.normal(0, 1, 100)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
poly_reg = Pipeline([
    ("poly", PolynomialFeatures(degree = 2)),
    ("std_scaler", StandardScaler()),
    ("lin_reg", LinearRegression())
])
```

```python
poly_reg.fit(X, y)
y_pred = poly_reg.predict(X)
```

```python
plt.scatter(x, y)
plt.plot(np.sort(x), y_pred[np.argsort(x)], color='r')
plt.show()
```
