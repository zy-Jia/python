```python
# split categorical columns and numerical columns
categorical_mask = (kc_data.dtypes == object)
categorical_cols = kc_data.columns[categorical_mask].tolist()
numeric_cols = kc_data.columns[~categorical_mask].tolist()
numeric_cols.remove('id')
numeric_cols.remove('price')
print(f'categorical_cols are {categorical_cols}' )
print(f'numeric_cols are {numeric_cols}' )
X = kc_data[numeric_cols]
X.head()
# X = kc_data
# X = X.drop(["yr_built", "id", "price", "date"], axis = 1)
```

```python
y = kc_data[['price']]
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```
